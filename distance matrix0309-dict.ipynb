```python
import pandas as pd
import numpy as np
df1=pd.read_csv('20134.csv',encoding = "GB2312") #import document
```

```python
from math import radians, cos, sin, asin, sqrt  
def haversine(lon1, lat1, lon2, lat2): # 经度1，纬度1，经度2，纬度2 （十进制度数）  
    # 将十进制度数转化为弧度  
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])   
    # haversine公式  
    dlon = lon2 - lon1   
    dlat = lat2 - lat1   
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2  
    c = 2 * asin(sqrt(a))   
    r = 6371.137 # 地球平均半径，单位为公里
    d=round(r*c, 3)
    return d 
```

```python
import time
start_time = time.time()

import numpy as np
np.set_printoptions(suppress=True)
#n=df1.shape[0]
n=500
dict={}
i=0
while i<n-1:
    a=np.empty(shape=[1, n-i-1])#创建空白矩阵
    j=i+1
    while j<n:
        lon1=df1.loc[i]['lng']
        lat1=df1.loc[i]['lat']
        lon2=df1.loc[j]['lng']
        lat2=df1.loc[j]['lat']
        a[0,j-1-i]=haversine(lon1,lat1,lon2,lat2)
        j=j+1
    a = np.float32(a)
    dict[i]=a
    i=i+1

print("--- %s seconds ---" % (time.time() - start_time))   
```

```python
import pickle
with open('data2.pickle','wb') as file:
    pickle.dump(dict,file)
```
